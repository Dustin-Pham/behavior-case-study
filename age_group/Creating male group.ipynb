```python
  import pandas as pd

print('pandas imported!')
```

```python
import os
working_directory = os.getcwd()
print(working_directory)
```

```python
path = working_directory + '/male_few_times.csv'
df=pd.read_csv(path)

print(df)
```

```python
maxValues = df.max()
print(maxValues)
```

```python
minValues = df.min()
print(minValues)
```

```python
# Split the column of age into three age groups
df["age_group"] = pd.cut(df["age"], bins=[15, 24, 54, 100], labels=["15-24", "25-54", "55-75"])

# Count the number of customers in each age group
age_group_counts = df.groupby("age_group").size()

# Add the age_group_counts Series to the existing table
df["age_group_counts"] = age_group_counts

print(df)
```

```python
# Save the DataFrame to a CSV file
df.to_csv("updated_male_few_times.csv")
```
